```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [1,3,4] 
y = [4,1,3]
plt.plot(x,y)
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
xpoint=np.array([1,3,4])
ypoint=np.array([4,1,3])
plt.plot(xpoint,ypoint,'o')
plt.grid()
plt.show()
```

```python
ypoints=np.array([4,1,3])

plt.plot(ypoints,marker='s')
plt.show()
```

```python
ypoints=np.array([4,1,3])
xpoints=np.array([1,4,3])

plt.plot(ypoints,xpoints,marker='o',ms=25,mec='b',mfc='y')
plt.grid()
plt.show()
```

```python
x=(1,4,3)
y=(3,4,1)
plt.subplot(1,2,1)
plt.plot(x,y,marker='o')
plt.show()
x=(1,2,3)
y=(4,5,6)
plt.subplot(1,2,2)
plt.plot(x,y,marker='d')
plt.show()
```

```python
#bar
# h bar 
deepu=['abc','bcd','cde']
mahi=[4,5,16]
plt.barh(deepu,mahi)
plt.title('love')
plt.xlabel('deepu')
plt.ylabel('mahi')
plt.show()
```

```python
#bar
# v bar 
deepu=['abc','bcd','cde']
mahi=[4,5,16]
plt.bar(deepu,mahi)
plt.title('love')
plt.xlabel('deepu')
plt.ylabel('mahi')
plt.show()
```

```python
plt.bar([1,3,5,7],[7,8,9,7],
label='duke',color='b',width=.5)
plt.bar([2,4,6,8],[4,2,3,6],
label='ninja',color='g',width=.5)
plt.legend()
plt.title('bikes')
plt.xlabel('units')
plt.ylabel('months')
plt.show()
```

```python
bb=[9234,4345,7616,7978,]
i=['india','uk','aus','uk',]
e=[0.5,0,0,0]
c=['red','orange', 'blue', 'green']

plt.pie(bb,labels=i,explode=e,shadow=True,startangle=90,wedgeprops={'edgecolor':'black', 'linewidth': 6 },colors=c)
plt.show()
```

```python
ypoints=np.array([4,1,3])
xpoints=np.array([1,4,3])

plt.scatter(ypoints,xpoints,color='r')
ypoints=np.array([1,2,3,7])
xpoints=np.array([1,4,3,4,6,9])
plt.scatter(ypoints,xpoints,color='b')

plt.show()
```
